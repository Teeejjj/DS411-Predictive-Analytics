```python
reader = easyocr.Reader(['en'])

result = reader.readtext('./Datasets/meta-image.png')
for (bbox,text,prob) in result:
    print(f'text: {text}, Probabilty: {prob}')
```

```python
texts = []
for (bbox,text,prob) in result:
    texts.append(text)
print(texts)
```

```python
im2 = Image.open('./Datasets/medium_article_sample.png')
display(im2.size,im2)
```

```python
result2 = reader.readtext('./Datasets/medium_article_sample.png')
for (bbox,text,prob) in result2:
    print(f'text: {text}, Probabilty: {prob}')
```

```python
text_list = []
for (_,text,_) in result2:
    text_list.append(text)
text_str = str(text_list)
text_str = ' '.join(text_str.split("',"))
cleaned_text = text_str.replace("[\'", "").replace("\']", "").replace("\'", "").replace("\n", "").replace("  ", " ")
cleaned_text = cleaned_text.split('.')
cleaned_text
```
